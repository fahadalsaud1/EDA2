```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sn
```

```
results = pd.read_csv('results.csv')
stats = pd.read_csv('stats.csv')
```

# Data Wrangling
Although my data wrangling wasn't too difficult, I had to remove certain columns that were missing data points in order to have a cleaner and more useful dataset.

```
stats.drop(["dispossessed", "big_chance_missed", "backward_pass", "backward_pass", "total_through_ball"], axis=1)
```

```
# Finding the use of home advantage

hw = results.query("home_goals > away_goals")
home_wins = (hw.shape[0])
aw = results.query("away_goals > home_goals")
away_wins = (aw.shape[0])
dr = results.query("away_goals == home_goals")
draw = (dr.shape[0])
value = [home_wins, away_wins, draw]
lab= 'Home Wins','Away Wins','Draws'
plt.pie(value, labels = lab, autopct='%1.1f%%')
plt.title('Home Advantage')
plt.axis('equal')
plt.show()
```

```
# This is the amount of goals per year
# My hypothesis on why this is happening is because overtime, teams tried score more goals in order to attract more fans. 
stats["Goals"] = stats['goals']
stats.groupby('season')["Goals"].sum().plot()
plt.title('Goals per Season')
plt.ylabel('Goals')
```

```
stats.groupby('team')["wins"].agg(['mean']).sort_values(ascending=False, by='mean').plot.bar(figsize = (10,5))
plt.title('Average Wins per Season')
plt.ylabel('Wins')
plt.xlabel('Teams')
lulu = stats.groupby('team')['wins'].mean()
# This graph shows the best teams, on average, from 2006-2018
```

```
yot = results[results.home_goals > results.away_goals]
hel = yot['home_team'].value_counts()
awt = results[results.away_goals > results.home_goals]
sec = awt['away_team'].value_counts()
frt = (hel/(sec + hel)).sort_values(ascending=False)
frt.plot.bar(figsize = (10,5))
plt.title('Percent of Wins coming at home')
plt.ylabel('% of wins at home')
plt.xlabel('Teams')
# This graph was used to measure the home field advantage of each team, and which teams relied on a home crowd for their victories.
```

```
topp = stats.loc[stats.groupby('season')['wins'].idxmax()]
secondw = topp.team.value_counts() 

seasonsinprem = stats['team'].value_counts()
ily = hel/seasonsinprem
ily
msy = sec/seasonsinprem

ily.corr(secondw)
# I tried to find a correlation between home wins and titles won vs away wins and titles won. 
# Since the home wins correlation was greater than the away wins correlation, teams that won the title typically win much more games at home than away.
```

```
msy.corr(secondw)
```

```
frt.corr(lulu)
# this correlation measured average wins and percentage of wins at home.
# This portrays that worse teams really on home field rather than the better teams.
```

This analysis should inspire teams to attack more (score more goals) and keep more possesion of the ball (amount of touches) if they want to win games. Also, there is a weak negative correlation between long balls and wins. Thus, although not a strong correlation, teams should try to limit playing long balls but rather rely on short passed.
