```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sn
```

```
results = pd.read_csv('results.csv')
stats = pd.read_csv('stats.csv')
```

```
stats.drop(["dispossessed", "big_chance_missed", "backward_pass", "backward_pass", "total_through_ball"], axis=1)
```

```
# Finding the use of home advantage

hw = results.query("home_goals > away_goals")
home_wins = (hw.shape[0])
aw = results.query("away_goals > home_goals")
away_wins = (aw.shape[0])
dr = results.query("away_goals == home_goals")
draw = (dr.shape[0])
value = [home_wins, away_wins, draw]
lab= 'Home Wins','Away Wins','Draws'
plt.pie(value, labels = lab, autopct='%1.1f%%')
plt.title('Home Advantage')
plt.axis('equal')
plt.show()
```

```
# This is the amount of goals per year
stats["Goals"] = stats['goals']
stats.groupby('season')["Goals"].sum().plot()
plt.title('Goals per Season')
plt.ylabel('Goals')
```

```
stats.groupby('team')["wins"].agg(['mean']).sort_values(ascending=False, by='mean').plot.bar(figsize = (10,5))
plt.title('Average Wins per Season')
plt.ylabel('Wins')
plt.xlabel('Teams')
```

```
yot = results[results.home_goals > results.away_goals]
hel = yot['home_team'].value_counts()
awt = results[results.away_goals > results.home_goals]
sec = awt['away_team'].value_counts()
frt = (hel/(sec + hel)).sort_values(ascending=False)
frt.plot.bar(figsize = (10,5))
plt.title('Percent of Wins coming at home')
plt.ylabel('% of wins at home')
plt.xlabel('Teams')
```
